```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# load and split data
df = pd.read_csv("spiral_chart_vals_orientation.csv")
df_data = df[df["Entry_Type"] == "SpiralData"]
df_guide = df[df["Entry_Type"] == "SpiralGuide"]
N = len(df_data)

num2month= {
    '01' : 'Jan',
    '02' : 'Feb',
    '03' : 'Mar',
    '04' : 'Apr',
    '05' : 'May',
    '06' : 'Jun',
    '07' : 'Jul',
    '08' : 'Aug',
    '09' : 'Sep',
    '10' : 'Oct',
    '11' : 'Nov',
    '12' : 'Dec'
}
# dates_to_keep = ["2019-01-01", "2020-01-01", "2021-01-01", "2022-01-01", "2023-01-01", "2024-01-01"]
# mask = ~df_data["Date"].isin(dates_to_keep)
# df_data.loc[mask, "Value_Pct"] = 0

# main plot figure
fig = go.Figure(

    data=[
          
        # SECTION 1 :: static guidelines and axes
        # --------------------------------------------

        # spiral
        go.Scatter(x=df_guide['x'], y=df_guide['y'],
            mode="lines", showlegend= False,
            line=dict(width=2, color="#a7a5a5")
            ),
        
        # dashed line to connect spiral and line graph
        go.Scatter(x=[df_guide['x'].values[-1], 600], y=[df_guide['y'].values[-1], df_guide['y'].values[-1]],
            mode="lines", showlegend= False,
            line=dict(dash= "longdash", width=2, color="#a7a5a5")
            ),
        
        # x axis
        go.Scatter(x=[600, 2_100], y=[df_guide['y'].values[-1], df_guide['y'].values[-1]],
            mode="lines", showlegend= False,
            line=dict(width=2, color="#a7a5a5")
            ),
        
        # 0% axis line label
        go.Scatter(
            x= [2_158],
            y= [df_guide['y'].values[-1]],
            mode='text',
            text=['0%'], 
            textfont=dict(size=18, color='#a7a5a5'), 
            showlegend=False),
        
        # -10% axis line and label
        go.Scatter(x=[550, 2_100], y=[df_guide['y'].values[-1]-100, df_guide['y'].values[-1]-100],
            showlegend= False, mode="lines",
            line=dict(width=2, color="#cfcdcd")),
        go.Scatter(
            x= [2_150],
            y= [df_guide['y'].values[-1]-100],
            mode='text',
            text=['-10%'], 
            textfont=dict(size=18, color='#cfcdcd'), 
            showlegend=False),
    
        # +10% axis line and label
        go.Scatter(x=[550, 2_100], y=[df_guide['y'].values[-1]+100, df_guide['y'].values[-1]+100],
            showlegend= False, mode="lines",
            line=dict(width=2, color="#cfcdcd")), 
        go.Scatter(
            x= [2_150],
            y= [df_guide['y'].values[-1]+100],
            mode='text',
            text=['+10%'], 
            textfont=dict(size=18, color='#cfcdcd'), 
            showlegend=False),

        # -20% axis line and label   
        go.Scatter(x=[500, 2_100], y=[df_guide['y'].values[-1]-200, df_guide['y'].values[-1]-200],
            showlegend= False, mode="lines",
            line=dict(width=2, color="#cfcdcd")),
        go.Scatter(
            x= [2_150],
            y= [df_guide['y'].values[-1]-200],
            mode='text',
            text=['-20%'], 
            textfont=dict(size=18, color='#cfcdcd'), 
            showlegend=False),
    
        # +20% axis line and label
        go.Scatter(x=[500, 2_100], y=[df_guide['y'].values[-1]+200, df_guide['y'].values[-1]+200],
            showlegend= False, mode="lines",
            line=dict(width=2, color="#cfcdcd")),
        go.Scatter(
            x= [2_150],
            y= [df_guide['y'].values[-1]+200],
            mode='text',
            text=['+20%'], 
            textfont=dict(size=18, color='#cfcdcd'), 
            showlegend=False),

        # --------------------------------------------


        # SECTION 2 :: moving elements (+ the circle)
        # --------------------------------------------

        # Q1 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"), showlegend= False,
            x= [450*np.cos(df_data.iloc[0]["Radians_Angle"]), 390*np.cos(df_data.iloc[0]["Radians_Angle"])],
            y= [450*np.sin(df_data.iloc[0]["Radians_Angle"]), 390*np.sin(df_data.iloc[0]["Radians_Angle"])]),
        go.Scatter(
            x= [500*np.cos(df_data.iloc[0]["Radians_Angle"])],
            y= [500*np.sin(df_data.iloc[0]["Radians_Angle"])],
            mode='text',
            text=['Q1'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),
        
        # Q2 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"), showlegend= False,
            x= [450*np.cos(df_data.iloc[92]["Radians_Angle"]), 390*np.cos(df_data.iloc[92]["Radians_Angle"])],
            y= [450*np.sin(df_data.iloc[92]["Radians_Angle"]), 390*np.sin(df_data.iloc[92]["Radians_Angle"])]),
        go.Scatter(
            x= [500*np.cos(df_data.iloc[92]["Radians_Angle"])],
            y= [500*np.sin(df_data.iloc[92]["Radians_Angle"])],
            mode='text',
            text=['Q2'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),

        # Q3 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"), showlegend= False,
            x= [-390*np.cos(df_data.iloc[0]["Radians_Angle"]), -450*np.cos(df_data.iloc[0]["Radians_Angle"])],
            y= [-390*np.sin(df_data.iloc[0]["Radians_Angle"]), -450*np.sin(df_data.iloc[0]["Radians_Angle"])]),
        go.Scatter(
            x= [-500*np.cos(df_data.iloc[0]["Radians_Angle"])],
            y= [-500*np.sin(df_data.iloc[0]["Radians_Angle"])],
            mode='text',
            text=['Q3'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),

        # Q4 tick and label    
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"), showlegend= False,
            x= [-390*np.cos(df_data.iloc[92]["Radians_Angle"]), -450*np.cos(df_data.iloc[92]["Radians_Angle"])],
            y= [-390*np.sin(df_data.iloc[92]["Radians_Angle"]), -450*np.sin(df_data.iloc[92]["Radians_Angle"])]),
        go.Scatter(
            x= [-500*np.cos(df_data.iloc[92]["Radians_Angle"])],
            y= [-500*np.sin(df_data.iloc[92]["Radians_Angle"])],
            mode='text',
            text=['Q4'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),
        
        # circle - !!! NOT DYNAMIC
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"), showlegend= False,
            x = [420 * np.cos(i * -(2 * np.pi / (365))) for i in range(366)],
            y = [420 * np.sin(i * -(2 * np.pi / (365))) for i in range(366)]),

        # year
        go.Scatter(
              x= [2_050],
              y= [df_guide['y'].values[-1]-370],
              mode='text',
              text=[f"{df_data['Date'].values[0].split('-')[0]}"], 
              textfont=dict(size=18, color='#a7a5a5'), 
              showlegend=False),
        
        # month
        go.Scatter(
              x= [2_050],
              y= [df_guide['y'].values[-1]-320],
              mode='text',
              text=[num2month[df_data['Date'].values[0].split('-')[1]]], 
              textfont=dict(size=16, color='#a7a5a5'), 
              showlegend=False),
        
        # datapoints  
        go.Scatter(x=df_data['x'], y=df_data['y'],
            showlegend= False,
            mode="markers",
            marker=dict(color=df_data['Value_Pct'].values, colorscale=[[0.0, "red"], [0.5, "lightgray"],[1.0, "green"]], 
                        cmin=-max(abs(df_data['Value_Pct'])), cmax=max(abs(df_data['Value_Pct'])), cmid=0,size=5))
        ])

        # --------------------------------------------

# create animation frames
frames = []
for n in range(1, N+1):
    
    # initialize data arrays
    x_vals = [None] * n
    y_vals = [None] * n

    # angles needed to properly plot circle elements
    angle_0 = df_data.iloc[0+30-(n%365)]["Radians_Angle"]
    angle_92 = df_data.iloc[92+30-(n%365)]["Radians_Angle"]

    years = []
    years_x = []

    month_1st_idx = []
    months = []
    months_x = []
    for idx in range(n):
        
        # last 30 points get a spot on the 'normal' axes, the rest get the spiral position based on their current step value and their percentage value

        if idx >= n-30:
            x_vals[idx] = 2_100 - ((n-idx) * 50)
            y_vals[idx] = df_guide['y'].values[-1] + (df_data['Value_Pct'].values[idx] * 10)

            year, month, day = df_data['Date'].values[idx].split('-')

            if year not in years:
                years.append(year)
                years_x.append(False)
            
            if num2month[month] not in months:
                months.append(num2month[month])
                months_x.append(False)

            if day == '01':
                months_x[-1] = 2_100 - ((n-idx) * 50)
                if month == '01':
                    years_x[-1] = 2_100 - ((n-idx) * 50)


        else:
            x_vals[idx] = (((df_data['Step'].values[-1]+30-(n-idx)) * 0.15) + (df_data['Value_Pct'].values[idx] * 1.5)) * np.cos((df_data['Step'].values[-1]+30-(n-idx)) * -(2 * np.pi / (365))-4.70)
            y_vals[idx] = (((df_data['Step'].values[-1]+30-(n-idx)) * 0.15) + (df_data['Value_Pct'].values[idx] * 1.5)) * np.sin((df['Step'].values[-1]+30-(n-idx)) * -(2 * np.pi / (365))-4.70)

    if len(months) == 1:
        months.append("X")
        months_x.append(2_300)
    if months_x[0] is False:
        months_x[0] = 600

    if len(years) == 1:
        years.append("X")
        years_x.append(2_300)
    if years_x[0] is False:
        years_x[0] = 600

    # add new frame to the list
    frames.append(go.Frame(
    name= df_data['Date'].values[n-1],
    data=[

        # Q1 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"),
            x= [450*np.cos(angle_0), 390*np.cos(angle_0)],
            y= [450*np.sin(angle_0), 390*np.sin(angle_0)]),
        go.Scatter(
              x= [500*np.cos(angle_0)],
              y= [500*np.sin(angle_0)],
              mode='text',
              text=['Q1'], 
              textfont=dict(size=24, color='#cfcdcd'), 
              showlegend=False),
        
        # Q2 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"),
            x= [450*np.cos(angle_92), 390*np.cos(angle_92)],
            y= [450*np.sin(angle_92), 390*np.sin(angle_92)]),
        go.Scatter(
            x= [500*np.cos(angle_92)],
            y= [500*np.sin(angle_92)],
            mode='text',
            text=['Q2'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),

        # Q3 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"),
            x= [-390*np.cos(angle_0), -450*np.cos(angle_0)],
            y= [-390*np.sin(angle_0), -450*np.sin(angle_0)]),
        go.Scatter(
            x= [-500*np.cos(angle_0)],
            y= [-500*np.sin(angle_0)],
            mode='text',
            text=['Q3'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),

        # Q4 tick and label
        go.Scatter(
            mode= "lines", line= dict( width= 2, color= "#cfcdcd"),
            x= [-390*np.cos(angle_92), -450*np.cos(angle_92)],
            y= [-390*np.sin(angle_92), -450*np.sin(angle_92)]),
        go.Scatter(
            x= [-500*np.cos(angle_92)],
            y= [-500*np.sin(angle_92)],
            mode='text',
            text=['Q4'], 
            textfont=dict(size=24, color='#cfcdcd'), 
            showlegend=False),
        
        # year
        go.Scatter(
              x= years_x,
              y= [df_guide['y'].values[-1]-370] * len(years),
              mode='text',
              text=years, 
              textfont=dict(size=18, color='#a7a5a5'), 
              showlegend=False),

        # month
        go.Scatter(
              x= months_x,
              y= [df_guide['y'].values[-1]-320] * len(months),
              mode='text',
              text=months, 
              textfont=dict(size=16, color='#a7a5a5'), 
              showlegend=False),      

        # data points
        go.Scatter(
            x=x_vals + [2_300],
            y=y_vals + [df_guide['y'].values[-1]],
            marker=dict(
                color=df_data['Value_Pct'].values[:n],
                colorscale=[[0.0, "red"], [0.5, "lightgray"], [1.0, "green"]],
                cmin=-max(abs(df_data['Value_Pct'])),
                cmax=max(abs(df_data['Value_Pct'])),
                cmid=0,
                size=5
        ))
    ],

    traces=[12, 13, 14, 15, 16, 17, 18, 19, 21, 22, 23]
    ))

# slider
slider_steps =[]
for n in range(1, N+1):
    step = dict(
        method="animate",
        args=[
            [df_data['Date'].values[n-1]],
            dict(
                mode="immediate",
                frame=dict(duration=50, redraw=True), 
                transition=dict(duration=0)
            )
        ],
        label=df_data['Date'].values[n-1]
    )
    slider_steps.append(step)

# update the layout
fig.update_layout(width=2_000, height=1_000,
        plot_bgcolor='white',
        paper_bgcolor='white',
        xaxis=dict(range=[min(df_data['x'].min(), df_guide['x'].min())-200, 2_200], autorange=False, zeroline=False, visible= False),
        yaxis=dict(range=[min(df_data['y'].min(), df_guide['y'].min())-200, max(df_data['y'].max(), df_guide['y'].max(), df_guide['y'].values[-1]+250)+200], autorange=False, zeroline=False, visible= False),
        title_text="Spiral test", title_x=0.5,

        # action buttons - duration controls animation speed
        updatemenus = [dict(type = "buttons", x=0.05, y=-0.08, direction= 'right',
        buttons = [
            dict(
                args = [None, {"frame": {"duration": 150, "redraw": False},
                                "fromcurrent": True, "transition": {"duration": 2_500}}],
                label = "Play",
                method = "animate"),

            dict(
                args = [[None], {"frame": {"duration": 0, "redraw": False},
                                "mode" : "immediate", "transition": {"duration": 0}}],
                label = "Pause",
                method = "animate")
                ])],

        sliders=[dict(
        active=0,
        x=0.08, y=-0.08,
        len=0.50,
        tickwidth=0,
        font=dict(color='rgba(0,0,0,0)'),
        currentvalue=dict(
            font=dict(size=16, color= 'black'),
            prefix="Date: ", 
            visible=True,
            xanchor="left"
        ),
        transition=dict(duration=2_500),
        steps=slider_steps)]        
                )

fig.frames = frames
fig.show()
```
