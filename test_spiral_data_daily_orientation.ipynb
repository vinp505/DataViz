```python
import pandas as pd
import numpy as np

# 1. Generate Weekly Dates: freq='W' gives us 1 row per week
dates = pd.date_range(start='2019-01-01', end='2025-12-31', freq='D')
steps = np.arange(len(dates))

# 2. Generate Percentage Data
np.random.seed(42)
# Since we have ~52.14 weeks in a year, we adjust the sine wave to match
base_seasonality = np.sin(steps * (2 * np.pi / 365)) * 10 
noise = np.random.normal(0, 3, len(dates))
percentages = np.round(base_seasonality + noise, 1)

df = pd.DataFrame({
    'Date': dates, 'Step': steps, 'Value_Pct': percentages
})

df
```

```python
df['Step'] = np.arange(len(df))
df.tail(10)
```

```python
df['Radians_Angle'] = (df['Step'] * -(2 * np.pi / (365))) -4.70
df
```

```python
df['Radius'] = (df['Step'] * 0.15) + (df['Value_Pct'] * 1.5)
df
```

```python
df['x'] = df['Radius'] * np.cos(df['Radians_Angle'])
df['y'] = df['Radius'] * np.sin(df['Radians_Angle'])
df['Entry_Type'] = "SpiralData"
```

```python
df_gl = df[["Step", "Radians_Angle"]].copy()
df_gl['Radius'] = df_gl['Step'] * 0.15
df_gl['x'] = df_gl['Radius'] * np.cos(df_gl['Radians_Angle'])
df_gl['y'] = df_gl['Radius'] * np.sin(df_gl['Radians_Angle'])
df_gl['Entry_Type'] = "SpiralGuide"
df_gl
```
